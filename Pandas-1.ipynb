# Indexing and Selecting data
pandas provides a suite of methods in order to get integer and label based indexing. The semantics follow closely python and numpy slicing

```python
# To get columns 
df[["Fare","Survived"]]
```

```python
#Get a cell
df["Fare"][3]
```

```python
# Get rows
df[2:6]
```

## .loc
This is for selection by label

dataframe.loc[row labels,column labels]

```python
df.loc[0:4,:]
```

```python
df.loc[0:5,"Name":"Ticket"]
```

```python
df_new = pd.DataFrame(np.random.randn(4,4), columns=list('ABCD'), index=list('abcd'))
df_new
df_new.loc["c":,"A":"C"]
```

```python
#Get Cell
df_new.loc["c","A"]
```

## .iloc
Selection by position

```python
df.iloc[2,:]
```

```python
df.iloc[2:5,0:3]
```

## .ix
Mixed indexing


```python
df_new = pd.DataFrame(np.random.randn(4,4), columns=list('ABCD'), index=list('abcd'))
df_new
df_new.ix[1:4,"A":"C"]
```

```python
df_new.ix["c":"d",0:]
```

## Boolean indexing

```python
df[(df["Age"]>50) & (df["Sex"]=="male")]
```

```python
df[(df["Fare"]>50 )| (df["Embarked"]=="S")]
```

# Group By: split-apply-combine

```python
df.groupby(["Pclass"]).groups.keys()
```

```python
df.groupby(["Pclass","Sex"]).groups.keys()
```

```python
df.groupby("Sex").get_group("male")
```

```python
df.groupby("Pclass").count()
```

```python
df.groupby("Pclass").mean()
```

```python
df.groupby("Pclass").agg(np.median)
```

```python
df.groupby("Pclass").agg([np.sum,np.mean])
```

```python
df.groupby("Pclass").agg({"Fare":lambda x : np.median(x),
                           "Age":np.median})
```
